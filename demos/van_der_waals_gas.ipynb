```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# Van der Waals Gas

The equation of state is

$(P + a \frac{N^2}{V^2})(V - Nb) = N k_B T$

```python
a = 1.
b = 0.01
N = 10.
kB = 1.

V = np.linspace(0.15, 0.8, 100)

TList = [20., 30., 40.]
for T in TList:
    P = N*kB*T/(V - N*b) - a*N**2/V**2
    plt.plot(V, P, label="T={}".format(T))

plt.legend()
plt.ylim([-500, 2000])
plt.xlabel("V")
plt.ylabel("P")
```
